```python

# Create a mapping of spotify_id to artist name
spotify_to_artist = dict(zip(nodes['spotify_id'], nodes['name']))

# Define a function to get the artist name
def get_artist_name(spotify_id):
    return spotify_to_artist.get(spotify_id, 'Unknown')  # 'Unknown' if not found

# Create artist_0 and artist_1 columns
edges['artist_0'] = edges['id_0'].apply(get_artist_name)
edges['artist_1'] = edges['id_1'].apply(get_artist_name)

# Show the resulting DataFrame
print(edges)
```

```python
edges.isnull().sum()
```

```python
null_rows = edges[edges['artist_0'].isnull() | edges['artist_1'].isnull()]

# Display rows with nulls
print(null_rows)
```

```python
# Remove rows where 'artist_0' or 'artist_1' is null
edges_cleaned = edges.dropna(subset=['artist_0', 'artist_1'])

# If you want to remove these rows from the original DataFrame, you can use the inplace parameter
edges.dropna(subset=['artist_0', 'artist_1'], inplace=True)

# Display the cleaned DataFrame
print(edges_cleaned)
```

```python
edges.isna().sum()
```

```python

# Remove the columns 'id_0' and 'id_1'
edges_cleaned = edges_cleaned.drop(columns=['id_0', 'id_1'])

# Export the updated DataFrame to a new CSV file
edges_cleaned.to_csv('collaborated_artist.csv', index=False)

# Verify by displaying the first few rows of the new DataFrame
print(edges_cleaned.head())
```
